```python
data = pd.read_csv('../input/ds-dataset/houses_train.csv')
data

### Генерация признаков

df=data.copy()
all_time=list(map(lambda x: (x[-6:-1]),np.array(df.date)))#оставили только само время, а не дату
all(np.array(all_time)=="00000")#как хорошо видно времени вообще нет
df.date=list(map(lambda x: str(x[0:8]),np.array(df.date)))
df.head()

year=list(map(lambda x: int(x[0:4]),np.array(df.date)))
month=list(map(lambda x: int(x[4:6]),np.array(df.date)))
day=list(map(lambda x: int(x[6:8]),np.array(df.date)))
df['date_year']=year
df['date_month']=month
date_season=(np.array(month))//3
date_season[date_season==4]*=0
df['date_season']=date_season
diff=list(df.apply(lambda x: np.abs(x.date_year-x.yr_built if x.yr_renovated==0 or x.yr_renovated>x.date_year else \
                                    x.date_year-x.yr_renovated), axis=1))
df['diff_yr_change_date']=diff
df.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)
```

Теперь давайте разберемся с категориальным признаком zipcode

```python
df_mean_encording=X_train.zipcode.value_counts().reset_index()
df_mean_encording.columns = ['zipcode','counts']
df_mean_encording['enconding']=list(map(lambda x: np.quantile(y_train[X_train.zipcode==x], q=0.5),df_mean_encording.zipcode))

b=np.sum(X_val.apply(lambda x: int(df_mean_encording[df_mean_encording.zipcode==x.zipcode].enconding), axis=1).isnull())
print(b)#проверили, что нет новых почтовых индексов, по которым мы обучались в val and test. Иначе надо было бы реализовывать данный случай
X_train['zipcode']=X_train.apply(lambda x: int(df_mean_encording[df_mean_encording.zipcode==x.zipcode].enconding), axis=1)
X_val['zipcode']=X_val.apply(lambda x: int(df_mean_encording[df_mean_encording.zipcode==x.zipcode].enconding), axis=1)
```

Будем находить всю информацию для одной модели и т.д. Это связано с тем, что для Catboost сам обработает категор. данные, для него будем использовать другие данные

**GradientBoostingRegressor**

```python
regressor = GradientBoostingRegressor(n_estimators=400, random_state=42)
regressor.fit(X_train, y_train)
predictions=regressor.staged_predict(X_val)
n_estimators_gr_boosting = [mape(y_val, p) for p in predictions]
n_estimators_gr_boosting_train = [mape(y_train, p) for p in regressor.staged_predict(X_train)]
```

```python
max_depth_gr_boosting=[]
max_depth_gr_boosting_train=[]
for max_depth in tqdm(range(2, 25)):
    regressor = GradientBoostingRegressor(max_depth=max_depth, 
                                      n_estimators=300, random_state=42)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    max_depth_gr_boosting.append(mape(y_val, predictions))
    max_depth_gr_boosting_train.append(mape(y_train, regressor.predict(X_train)))
```

```python
learning_rate_gr_boosting=[]
learning_rate_gr_boosting_train=[]
grid=np.logspace(-7,0,20)
for learning_rate in tqdm(grid):
    regressor = GradientBoostingRegressor(learning_rate=learning_rate,n_estimators=300, random_state=42)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    learning_rate_gr_boosting.append(mape(y_val, predictions))
    learning_rate_gr_boosting_train.append(mape(y_train, regressor.predict(X_train)))
```

**XGBoost**

```python
n_estimators=400
regressor = XGBRegressor(n_estimators=400, random_state=42, eval_metric='mape').fit(X_train, y_train)
n_estimators_XGBRegressor = [
    mape(
        y_val, regressor.predict(X_val, ntree_limit=i+1)
    )
    for i in range(n_estimators)
]

n_estimators_XGBRegressor_train = [
    mape(
        y_train, regressor.predict(X_train, ntree_limit=i+1)
    )
    for i in range(n_estimators)
]
```

```python
max_depth_XGBRegressor=[]
max_depth_XGBRegressor_train=[]
for max_depth in tqdm(range(2, 25)):
    regressor = XGBRegressor(n_estimators=300, random_state=42, eval_metric='mape',max_depth=max_depth).fit(X_train, y_train)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    max_depth_XGBRegressor.append(mape(y_val, predictions))
    max_depth_XGBRegressor_train.append(mape(y_train, regressor.predict(X_train)))
```

```python
learning_rate_XGBRegressor=[]
learning_rate_XGBRegressor_train=[]
grid=np.logspace(-7,0,20)
for learning_rate in tqdm(grid):
    regressor = XGBRegressor(n_estimators=300, random_state=42, eval_metric='mape',learning_rate=learning_rate).fit(X_train, y_train)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    learning_rate_XGBRegressor.append(mape(y_val, predictions))
    
    predictions = regressor.predict(X_train)
    learning_rate_XGBRegressor_train.append(mape(y_train, predictions))
```

**LightGBM**

```python
n_estimators=400
regressor = LGBMRegressor(n_estimators=400, random_state=42,metric='mape').fit(X_train, y_train)
n_estimators_LGBMRegressor = [
    mape(
        y_val, regressor.predict(X_val, num_iteration=i+1)
    )
    for i in tqdm(range(n_estimators))
]

n_estimators_LGBMRegressor_train = [
    mape(
        y_train, regressor.predict(X_train, num_iteration=i+1)
    )
    for i in tqdm(range(n_estimators))
]
```

```python
max_depth_LGBMRegressor=[]
max_depth_LGBMRegressor_train=[]
for max_depth in tqdm(range(2, 25)):
    regressor = LGBMRegressor(n_estimators=300, random_state=42, metric='mape',max_depth=max_depth).fit(X_train, y_train)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    max_depth_LGBMRegressor.append(mape(y_val, predictions))
    
    predictions = regressor.predict(X_train)
    max_depth_LGBMRegressor_train.append(mape(y_train, predictions))
```

```python
learning_rate_LGBMRegressor=[]
learning_rate_LGBMRegressor_train=[]
grid=np.logspace(-7,0,20)
for learning_rate in tqdm(grid):
    regressor = LGBMRegressor(n_estimators=300, random_state=42, metric='mape',learning_rate=learning_rate).fit(X_train, y_train)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    learning_rate_LGBMRegressor.append(mape(y_val, predictions))
    
    predictions = regressor.predict(X_train)
    learning_rate_LGBMRegressor_train.append(mape(y_train, predictions))
```

**CatBoost**

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)
cat_features_index=np.where(X_train.columns=='zipcode')[0]
```

```python
train_pool = Pool(X_train, label=y_train, cat_features=cat_features_index, 
                  feature_names=list(X_train.columns))
val_pool = Pool(X_val, label=y_val, cat_features=cat_features_index, 
                 feature_names=list(X_val.columns))
```

```python
n_estimators=400
regressor = CatBoostRegressor(n_estimators=400, random_state=42,eval_metric='MAPE', verbose=0).fit(train_pool, eval_set=val_pool)

n_estimators_CatBoostRegressor = [
    mape(
        y_val, regressor.predict(X_val, ntree_end=i+1)
    )
    for i in tqdm(range(n_estimators))
]

n_estimators_CatBoostRegressor_train = [
    mape(
        y_train, regressor.predict(X_train, ntree_end=i+1)
    )
    for i in tqdm(range(n_estimators))
]
```

```python
max_depth_CatBoostRegressor=[]
max_depth_CatBoostRegressor_train=[]
for max_depth in tqdm(range(2, 16)):
    regressor = CatBoostRegressor(n_estimators=300, random_state=42, eval_metric='MAPE',depth=max_depth,verbose=0).fit(train_pool, eval_set=val_pool)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    max_depth_CatBoostRegressor.append(mape(y_val, predictions))
    
    predictions = regressor.predict(X_train)
    max_depth_CatBoostRegressor_train.append(mape(y_train, predictions))
```

```python
learning_rate_CatBoostRegressor=[]
learning_rate_CatBoostRegressor_train=[]
grid=np.logspace(-6.9,0,20)
for learning_rate in tqdm(grid):
    regressor = CatBoostRegressor(n_estimators=300, random_state=42, eval_metric='MAPE',
                                  learning_rate=learning_rate,verbose=0).fit(train_pool, eval_set=val_pool)
    regressor.fit(X_train, y_train)
    predictions = regressor.predict(X_val)
    learning_rate_CatBoostRegressor.append(mape(y_val, predictions))
    
    predictions = regressor.predict(X_train)
    learning_rate_CatBoostRegressor_train.append(mape(y_train, predictions))
```
